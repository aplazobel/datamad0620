#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
#your code-1st way
sns.jointplot("fare","age", data=titanic, kind="reg")
```

```python
# your code-2nd way
plt.scatter(titanic.fare,titanic.age)
plt.xlabel('Fare')
plt.ylabel('Age')
m, b = np.polyfit(titanic.fare,titanic.age, 1)
plt.plot(titanic.fare, m*titanic.fare + b)
```

#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
#your code here
sns.heatmap(titanic.select_dtypes(include=['int64','float64']),vmin=0, vmax=1)

corr = titanic.select_dtypes(include=['int64','float64']).corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
